## Quick Look at the Data

The starter dataset contains anonymized web pages with search performance metrics. The statistics below provide evidence that the dataset is suitable for studying content refresh prioritization. The total number of pages, average search volume, average CTR, and the number of declining pages indicate that the dataset contains enough information to explore meaningful patterns.

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Total Pages:", len(df))
print("Average Search Volume:", round(df["search_volume"].mean(), 2))
print("Average CTR:", round(df["ctr"].mean(), 4))
print("Declining Pages:", (df["trend_direction"] == "down").sum())
```
